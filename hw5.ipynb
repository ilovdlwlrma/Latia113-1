```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

dates = pd.date_range("20130101", periods=100)
df = pd.DataFrame(np.random.rand(100, 4), index=dates, columns=list("ABCD"))
colors_A = plt.cm.Reds(np.linspace(0.3, 1, len(df)))
colors_B = plt.cm.Blues(np.linspace(0.3, 1, len(df)))
colors_C = plt.cm.Greens(np.linspace(0.3, 1, len(df)))
colors_D = plt.cm.Purples(np.linspace(0.3, 1, len(df)))
color_maps = [colors_A, colors_B, colors_C, colors_D]

fig, axes = plt.subplots(2, 2, figsize=(14, 14))
for i, col in enumerate(df.columns):
    ax = axes[i // 2, i % 2]
    ax.pie(df[col], labels=df[col].round(2), colors=color_maps[i], 
           autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.3))
    ax.set_title(f'{col} pie chart', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
colors = ['r', 'b', 'g', 'y']
fig, axes = plt.subplots(2, 2, figsize=(14, 14), subplot_kw={'projection': '3d'})
for i, col in enumerate(df.columns):
    ax = axes[i // 2, i % 2] 
    x, y, z = np.arange(len(df)), np.zeros(len(df)), np.zeros(len(df))
    dz = df[col].values
    ax.bar3d(x, y, z, dx=1, dy=1, dz=dz, color=colors[i], zsort='average')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy import stats
group1 = [85, 86, 88, 75, 78, 94]
group2 = [92, 94, 89, 96, 93, 95, 76, 50]
group3 = [78, 81, 79, 84, 77, 80]
f_statistic, p_value = stats.f_oneway(group1, group2, group3)
print("F 統計量:", f_statistic)
print("p-value:", p_value)

alpha = 0.05
if p_value < alpha:
	print("拒絕原假設：不同組別之間的均值存在顯著差異。")
else:
	print("未能拒絕原假設：不同組別之間的均值沒有顯著差異。")
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
data = {'Score': np.random.randint(0, 101, 100),
		'Method': np.random.choice(['A', 'B', 'C'], 100)}

df = pd.DataFrame(data)
model = ols('Score ~ C(Method)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
plt.figure(figsize=(10, 6))
avg_scores = df.groupby('Method')['Score'].mean()
avg_scores.plot(kind='bar', color=['red', 'green', 'blue'])
plt.xlabel('Method')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.show()
```
