```python
import requests
import json
import pandas as pd

url = 'https://tcgbusfs.blob.core.windows.net/dotapp/youbike/v2/youbike_immediate.json'
web = requests.get(url)
data = json.loads(web.content.decode('utf-8'))
df = pd.DataFrame(data)
output_path = r'./youbike.csv'
df.to_csv(output_path, index=False, encoding='utf-8', errors='replace')

print(f"CSV output to {output_path}")
```

```python
import pandas as pd

df = pd.read_csv('youbike.csv')
print("Youbike2.0")

try:
    from IPython.display import display
    display(df)
except ImportError:
    print(df)
```

```python
import folium
from folium.plugins import HeatMap
from IPython.display import display

df['usage_rate'] = df['available_rent_bikes'] / (df['available_rent_bikes'] + df['available_return_bikes'])

m = folium.Map(location=[25.0330, 121.5654], zoom_start=13)
for index, row in df.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=8,
        color='green' if row['usage_rate'] < 0.5 else 'red',
        fill=True,
        fill_color='green' if row['usage_rate'] < 0.5 else 'red',
        fill_opacity=0.6,
        popup=f": {row['sna']}<br>: {row['usage_rate']:.2f}"
    ).add_to(m)

df_cleaned = df.dropna(subset=['latitude', 'longitude', 'usage_rate'])
heat_data = [[row['latitude'], row['longitude'], row['usage_rate']] for index, row in df_cleaned.iterrows()]
HeatMap(heat_data).add_to(m)
display(m)
```

```python
from bs4 import BeautifulSoup
import requests as rq
url = 'https://water.taiwanstat.com/'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.146 Safari/537.36'}
response = rq.get(url, headers = headers, timeout=2)
html = response.content
soup = BeautifulSoup(html, "html.parser") # create a new bs4 object from the html data loaded
print(soup)
```

```python
soup = BeautifulSoup(html, "html.parser")
for script in soup(["script", "style"]):
	script.extract()

text = soup.get_text()
lines = (line.strip() for line in text.splitlines())
chunks = (phrase.strip() for line in lines for phrase in line.split(" "))
texts = '\n'.join(chunk for chunk in chunks if chunk)
print(texts)
```

```python
import requests
from bs4 import BeautifulSoup
import os

save_dir = './Beauty/'
if not os.path.exists(save_dir):
	os.makedirs(save_dir)

web = requests.get('https://www.ptt.cc/bbs/Beauty/M.1724575490.A.2AA.html', cookies={'over18':'1'})
soup = BeautifulSoup(web.text, "html.parser")
imgs = soup.find_all('img')
name = 0
for i in imgs:
	print(i['src'])
	jpg = requests.get(i['src'])
	with open(f'{save_dir}test_{name}.jpg', 'wb') as f:
		f.write(jpg.content)
	name += 1
```
